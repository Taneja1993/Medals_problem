```python
# creating masks for all the four conditions checked above.
```

```python
mask1 = summer.Event.str.lower().str.contains("mixed")
mask2 = summer.Event.str.lower().str.contains("pairs")
mask3 = summer.Sport.str.lower().str.contains("equestrian")
mask4 = ((summer.Sport.str.lower().str.contains("sailing")) & (summer.Year < 1988))
```

```python
# This statement assigns the value 'X' to gender column cells where any of the above conditions met
# because satisfying any of the above conditions signify the entry of redundant medals in the table.
summer.loc[mask1 | mask2 | mask3 | mask4, "Event_Gender"] = "X"
```
